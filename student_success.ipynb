```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, precision_score
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
df = pd.read_csv("student_performance_interactions.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
lab = LabelEncoder()
df["grade"] = lab.fit_transform(df["grade"])
df["study_environment"] = lab.fit_transform(df["study_environment"])
```

```python
df["parent_education_level"].value_counts()
```

```python
df["parent_education_level"] = df["parent_education_level"].map({"High School":1, "Bachelor":2, "Master":3})
```

```python
df.head()
```

```python
df.info()
```

```python
X = df.drop(["pass_fail","student_id"], axis=1)
```

```python
y= df["pass_fail"]
```

```python
X_train,X_test, y_train,y_test = train_test_split(X,y, train_size=0.70, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
random_model = RandomForestClassifier()
random_model.fit(X_train,y_train)
y_predict = random_model.predict(X_test)
                                    
```

```python
accuracy_score(y_test, y_predict)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression()
log_model.fit(X_train,y_train)
log_y_predict = log_model.predict(X_test)
```

```python
accuracy_score(y_test,log_y_predict)
```
